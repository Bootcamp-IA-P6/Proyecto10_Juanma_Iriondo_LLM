# Probando modelos de ollama

## Importando librerias

```python
from ollama import chat
```

```python
# modelos en local con ollama
# llama3.2:1b (6.3 sg) , llama3:latest (11 sg), qwen2.5-coder:7b (15 sg)

response = chat(
    model='llama3.2:1b',
    messages=[{'role': 'user', 'content': 'Hola!'}],
)
print(response.message.content)
```

```python
numero_palabras = 50
plataforma = "instagram"
tema = "python"
numero_hashtags = 3

# pregunta = "Crea un artículo para instagram con 50 palabras sobre python"
pregunta = f"Crea un artículo para {plataforma} con {numero_palabras} palabras sobre {tema} con {numero_hashtags}"

response = chat(
    model='llama3.2:1b',
    messages=[{'role': 'user', 'content': pregunta}],
)
print(response.message.content)
```
