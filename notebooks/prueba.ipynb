## Cadenas con Prompt Template

```python
template = "Eres un asistente útil que traduce del {idioma_entrada} al {idioma_salida} el texto: {texto}."
texto = "Me encanta programar"
prompt_template = PromptTemplate(
    input_variables = ["idioma_entrada", "idioma_salida", "texto"], template=template
)
```

```python
# llamando al LLM
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
```

```python
llm = ChatGroq(temperature=0, model_name="llama-3.3-70b-versatile")
```

```python
# Creación de la cadena usando el operador LCEL (|)
cadena = prompt_template | llm

# Ejecución de la cadena
respuesta = cadena.invoke(input={"idioma_entrada": "español", "idioma_salida": "inglés", "texto": texto})
print(respuesta.content)
```

```python
# Prueba para que el modelo sea conciso y responda en una línea
modelo = ChatGroq(temperature=0, model_name="llama-3.3-70b-versatile", max_tokens=50)

texto = "Me gusta pintar cuadros al oleo"

# Indicamos en la instrucción del sistema el límite de palabras
prompt = ChatPromptTemplate.from_messages([
    #("system", "Eres un asistente que responde de forma muy breve y al grano, en un máximo de dos frases."),
    ("system", "Eres un asistente que responde de forma muy breve y al grano, en un máximo de una frase."),
    ("user", "Traduce de {idioma_entrada} a {idioma_salida} el {texto}")
])

# Creación de la cadena usando el operador LCEL (|)
cadena = prompt | modelo

# Ejecución de la cadena
respuesta = cadena.invoke({"idioma_entrada": "español", "idioma_salida": "inglés", "texto": texto})
print(respuesta.content)
```

### Forma de limpiar la salida

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
modelo = ChatGroq(temperature=0, model_name="llama-3.3-70b-versatile", max_tokens=50)

# Creación de la cadena usando el operador LCEL (|)
cadena_final = prompt | modelo | StrOutputParser()

# Ejecución de la cadena
respuesta_final = cadena_final.invoke({"idioma_entrada": "español", "idioma_salida": "inglés", "texto": texto})
print(respuesta_final)
```
